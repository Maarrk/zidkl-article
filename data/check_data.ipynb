```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(13, 7))
```

```python
data = np.loadtxt('log_20141114T153149.csv', delimiter=';', skiprows=1)
print(data.shape)
```

```python
count = 1800
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time [s]')
ax1.set_ylabel('load factor', color=color)
ax1.plot(data[:count, 0], data[:count, 1], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('airspeed [km/h]', color=color)  # we already handled the x-label with ax1
ax2.plot(data[:count, 0], data[:count, 2], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
